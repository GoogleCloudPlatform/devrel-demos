##  Compare and optimize prompt template design for Prompt Engineering

The Vertex AI SDK streamlines the structured compilation of text prompts from datasets and provides comprehensive evaluation results of multiple Prompt Templates for a specified evaluation task that can be compared for prompt engineering.

### Define an evaluation dataset


To perform pointwise inference, the evaluation dataset is required to contain the following fields:

* Instruction: Part of the input user prompt. It refers to the inference instruction that is sent to your LLM.
* Context: User input for the Gen AI model or application in the current turn.
* Reference: The ground truth to compare your LLM response to.

Your dataset must include a minimum of one evaluation example. We recommend around 100 examples to ensure high-quality aggregated metrics and statistically significant results.

### Define prompt templates to compare


```
prompt_templates = [
    "Instruction: {instruction}. First provide a one-sentence reason for your decision, then end with the label SPAM or LEGITIMATE. Social media post: '{context}'.",
    "Message: {context}. Complete this task: {instruction}."
]
```

### Define EvalTask & Experiment

```
generation_config = {
    "temperature": 0.3,
}

safety_settings = {
    HarmCategory.HARM_CATEGORY_UNSPECIFIED: HarmBlockThreshold.BLOCK_NONE,
    HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE,
    HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
    HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_NONE,
    HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_NONE,
}

gemini_model = GenerativeModel(
    "gemini-2.5-flash",
    generation_config=generation_config,
    safety_settings=safety_settings,
)
```

```
instruction = "Evaluate if this message is spam."

context = [
    "Just finished a fantastic hike! The views were breathtaking. #nature #adventure",
    "üî• LIMITED TIME OFFER! Get 50% off our new weight loss supplement! Link in bio! #fitness #sale",
    "Exciting news! Check out my latest project. Link in bio.",
    "You've been selected for a special offer! Reply YES to claim.",
    "Don't miss out on our exclusive webinar next week. Register now!",
    "Your package is delayed. Click here for updates.",
    "Earn passive income from home! DM me for details.",
    "Your account has been accessed from a new device. Verify your identity here.",
    "Congratulations! You've won a free vacation. Claim your prize now!",
    "New policy update regarding your data. Read more.",
    "Just shared some thoughts on productivity. Link in comments."
]

reference = [
    "The message describes a personal experience and contains no promotional content, suspicious links, or unsolicited offers. LEGITIMATE",
    "The message exhibits characteristics of unsolicited commercial advertising, including aggressive marketing language, a generic call to action, and promotion of a product often associated with spam. SPAM",
    "The message is generic, lacks personalization, and its sole purpose appears to be driving clicks to an unknown link, which are common characteristics of spam. SPAM",
    "This message uses vague language and a generic call to action common in unsolicited marketing attempts without providing any specific details about the offer or sender. SPAM",
    "The message is a generic marketing invitation lacking specific details or malicious indicators often found in spam. LEGITIMATE",
    "The message is highly suspicious because it lacks any specific package details and pressures the recipient to click an unverified link, a common characteristic of phishing attempts. SPAM",
    "This message uses common bait phrases for unrealistic financial gain and directs users to a private conversation for details, which are classic characteristics of spam or scam attempts. SPAM",
    "The message creates urgency and directs you to an external link for identity verification, which is a common phishing tactic. SPAM",
    "The message is spam because it offers an unsolicited, high-value prize with no context or identifying information, and demands immediate action, which are common characteristics of phishing or scam attempts. SPAM",
    "The message is highly generic and lacks specific identifying information about the sender or the recipient, which is a common characteristic of phishing attempts or unsolicited messages. SPAM",
    "The message is generic, lacks context about the sender or recipient, and is solely designed to drive clicks to an external link, which is characteristic of unsolicited promotional content. SPAM"
]

eval_dataset = pd.DataFrame(
    {
        "context": context,
        "reference": reference,
        "instruction": [instruction] * len(context),
    }
)
```

```
experiment_name = "eval-sdk-prompt-engineering"  # @param {type:"string"}
metrics = [
    "rouge_1",
    "rouge_l_sum"
]
summarization_eval_task = EvalTask(
    dataset=eval_dataset,
    metrics=metrics,
    experiment=experiment_name,
)
```

### Running Evaluation

```
run_id = notebook_utils.generate_uuid(8)
eval_results = []


for i, prompt_template in enumerate(prompt_templates):
    experiment_run_name = f"eval-prompt-engineering-{run_id}-prompt-{i}"

    eval_result = summarization_eval_task.evaluate(
        prompt_template=prompt_template,
        experiment_run_name=experiment_run_name,
        model=gemini_model,
    )

    eval_results.append((f"Prompt #{i}", eval_result))
```

### Display Evaluation report and explanations

```
for title, eval_result in eval_results:
    notebook_utils.display_eval_result(title=title, eval_result=eval_result)
```

```
for title, eval_result in eval_results:
    notebook_utils.display_explanations(eval_result, metrics=["summarization_quality"])
```

### Compare Eval Results

```
notebook_utils.display_bar_plot(
    eval_results,
    metrics=metrics,
)
```
